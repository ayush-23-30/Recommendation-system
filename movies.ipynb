```python
import ast 

def convert(obj):
  list = []
  for i in ast.literal_eval(obj): ## obj ka first value ko
    list.append(i['name']) # i ke ander id or name hai hum bss name dedo 
  return list;
```

```python
def get_top_actors(obj):
  list = []
  counter = 0
  for i in ast.literal_eval(obj): ## obj ka first value ko
    if counter != 3:
      list.append(i['name']) # i ke ander id or name hai hum bss name dedo 
      counter+=1
    else : break;
  return list 
```

```python
movies['cast'] = movies['cast'].apply(get_top_actors)
```

### Vectorization 
Converting a column `can be a category or regession` into a Vector `2D for now` 
- what is 2D vector?
it just two values showed on a graph  like- (5,2) ; (8,4) ; (3,1) etc first is the value of x and second is the value of y. 

- In our case we have 5000 movies; and every movie is a vector, show we suggest the `closest`one to the current vector 


#### Task - Convert task to vectors [Text Vectorization]
- we have many techniques - Word Embeddings, Bags-of-words,Term Frequency-Inverse Document Frequency (TF-IDF) etc
- ` Bag of Words ` -> this method represents text as the `frequency of words in a document`, ignoring grammar and word order.
- we will combine all the words , Then the most used words [`frequency word`].
- example - total word count 1 lakh,and most coming word is 5000 times then we will take that word out 
- we will do this for all the tags and we will find that, for one is (5000, 2) - 5000 movies and 2 most common . 2 is taken for a example 
- we will remove stop words - these are the words that are grammartic words like - are , is , i , like etc

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem (text):
  y = []
  for i in text.split():
    y.append(ps.stem(i));
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray(); 
```

```python
vectors
```

```python
vectors[0]
```

#### now, need to calculate the distance from another - to predict the movie 
-  more the distance, less the simiarites 
- we will find the `cosine diistance` 
- this checks the angle between all the vectors; the more  the angle;the less they are similiar 
- cos distance = 1 - cosine similarity


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)

```

```python
similarity[0] ## this is telling the score of similarity of one to remaining all ; that's why this is fiirst movie and the first score is 1 because its the same movie  similarity is 100% 

## we have total 4806 movies that's why the shape of this similarity is also 4806 ; its count similarity for all the movies 
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x :x[1])
## we reverse the sorted list beacuse all the similarity are in the last ; 
## enumerate give the index and turn it into a tuple 
## so, now i have every index for my similarity array 
## kon kis se kitna similar sab hai mere pass 
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x :x[1])[1:6]
## lambda function to get the key not the index sorted, i want the similar key that present in the [1] index
## [1:6] first se leke 6 tak dedo [top 5 ]
```

```python
def recommeded (movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x :x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
    # print(i[0]);
```

```python
recommeded('Avatar') ## return the top 5 movies 
```
